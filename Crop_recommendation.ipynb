```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import classification_report
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, roc_auc_score
```

```python
dataset = pd.read_csv('Crop_recommendation.csv')
```

```python
unique_crops = dataset['label'].unique()
print("Unique crop types :")
for crop in unique_crops:
  print(crop)
```

```python
from sklearn.preprocessing import LabelEncoder

# Extract the crop labels column from the dataset
crop_labels = dataset['label']

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform label encoder on crop labels
encoded_labels = label_encoder.fit_transform(crop_labels)

# Assign the encoded labels back to the last column of the dataset
dataset.iloc[:, -1] = encoded_labels


# Create a DataFrame with 'label' and 'Encoded_Crop_Labels' columns
encoded_df = pd.DataFrame({'label': crop_labels, 'Encoded_Crop_Labels': encoded_labels})

# Print unique values of 'label' and 'Encoded_Crop_Labels' columns
print(encoded_df.drop_duplicates())

```

```python
dataset.head(100)
```

```python
X = dataset.iloc[:,:7].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size= 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 3, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
RandomForestClassifier(criterion='entropy', n_estimators=3, random_state=0)
# from sklearn.tree import DecisionTreeClassifier
# classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
# classifier.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[105,	14,	50,	26.2148837,	87.6883982,	6.419052193,	59.65590798	]])))
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
