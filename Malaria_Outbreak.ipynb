```python
import numpy as np
dataset='outbreak_detect.csv'
import pandas as pd
df=pd.read_csv(dataset)
```

```python
#Label encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder

label_encoder= LabelEncoder()
df['Outbreak']=label_encoder.fit_transform(df['Outbreak'])

X=df.drop(['Outbreak'],axis=1)
Y=df['Outbreak']
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf=KNeighborsClassifier(n_neighbors=3)
knn_clf.fit(X_train,Y_train)
```

```python
y_pred=knn_clf.predict(X_test)
```

```python
Y_test
```

```python
# Convert y_pred and Y_test to NumPy arrays
# y_pred = y_pred.to_numpy()
# Y_test = Y_test.to_numpy()

# Reshape y_pred and Y_test
y_pred = y_pred.reshape(len(y_pred), 1)
Y_test = Y_test.reshape(len(Y_test), 1)

# Concatenate y_pred and Y_test
result = np.concatenate((y_pred, Y_test), axis=1)
print(result)
```
